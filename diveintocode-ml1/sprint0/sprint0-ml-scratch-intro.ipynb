# 【問題1】train_test_splitのスクラッチ
### まずはスクラッチの練習として、scikit-learnのtrain_test_splitを自作してみましょう。Jupyter Notebookでコーディングを進め、完成後はpyファイルとします。utilsディレクトリの中にsplit.pyを作ってください。

```python
from sklearn.datasets import load_iris
from sklearn import datasets
import pandas as pd
import random
import numpy as np

#データセット読み込み
iris = datasets.load_iris()

#データセットの準備 ラベルが０または１のみのデータを抜き出す。
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df_2 = df[df['target'] != 2]
display(df_2)

X = np.array(df_2.iloc[:,0:-1])
y = df_2.iloc[:,4]

#確認
display(X)
display(y)
```

```python
def train_test_split(X, y, train_size=0.8):
    """
    学習用データを分割する。

    Parameters
    ----------
    X : 次の形のndarray, shape (n_samples, n_features)
      学習データ
    y : 次の形のndarray, shape (n_samples, )
      正解値
    train_size : float (0<train_size<1)
      何割をtrainとするか指定

    Returns
    ----------
    X_train : 次の形のndarray, shape (n_samples, n_features)
      学習データ
    X_test : 次の形のndarray, shape (n_samples, n_features)
      検証データ
    y_train : 次の形のndarray, shape (n_samples, )
      学習データの正解値
    y_test : 次の形のndarray, shape (n_samples, )
      検証データの正解値
    """
    
    sample_size = X.shape[0]
    sample_index_list = list(range(0,sample_size))
    train_index_list = random.sample(sample_index_list, int(sample_size*train_size))
    test_index_list = list(set(sample_index_list) - set(train_index_list))
    X_train, X_test, y_train, y_test = X[train_index_list], X[test_index_list], y[train_index_list], y[test_index_list]
    

    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
display(X_train, X_test, y_train, y_test)
```

### ①irisデータセット

```python
#データセット読み込み
iris = datasets.load_iris()
#データセットの準備 ラベルが０または１のみのデータを抜き出す。
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df_2 = df[df['target'] != 2]
#X,yに格納
X = np.array(df_2.iloc[:,0:-1])
y = df_2.iloc[:,4]
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

<ロジスティック回帰>

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

#パイプラインの作成
pipe_logistic = Pipeline([('scl',StandardScaler()),('est',LogisticRegression(random_state=1))])
pipe_logistic.fit(X_train,y_train)

#予測する
y_pred = pipe_logistic.predict(X_test)

#スコアを算出
print("accuracy_score:",accuracy_score(y_test, y_pred))
print("precision_score:",precision_score(y_test, y_pred))
print("recall_score",recall_score(y_test, y_pred))
print("f1_score",f1_score(y_test, y_pred))
```

＜SVM＞

```python
from sklearn.svm import SVC

#パイプラインの作成
pipe_svc = Pipeline([('scl',StandardScaler()),('est',SVC(random_state=1))])
pipe_svc.fit(X_train,y_train)


#予測する
y_pred = pipe_svc.predict(X_test)

#スコアを算出
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
print("accuracy_score:",accuracy_score(y_test, y_pred))
print("precision_score:",precision_score(y_test, y_pred))
print("recall_score",recall_score(y_test, y_pred))
print("f1_score",f1_score(y_test, y_pred))
```

＜決定木＞

```python
from sklearn.tree import DecisionTreeClassifier

#パイプライン作成
pipe_tree = Pipeline([('scl',StandardScaler()),('est',DecisionTreeClassifier(random_state=1))])
pipe_tree.fit(X_train,y_train)


#予測する
y_pred = pipe_tree.predict(X_test)

#スコアを算出
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
print("accuracy_score:",accuracy_score(y_test, y_pred))
print("precision_score:",precision_score(y_test, y_pred))
print("recall_score",recall_score(y_test, y_pred))
print("f1_score",f1_score(y_test, y_pred))
```

### ②シンプルデータセット１

```python
import numpy as np

np.random.seed(seed=0)
n_samples = 500
f0 = [-1, 2]
f1 = [2, -1]
cov = [[1.0,0.8], [0.8, 1.0]]

f0 = np.random.multivariate_normal(f0, cov, int(n_samples/2))
f1 = np.random.multivariate_normal(f1, cov, int(n_samples/2))

X = np.concatenate((f0, f1))
y = np.concatenate((np.ones((int(n_samples/2))), np.ones((int(n_samples/2))) *(-1))).astype(np.int)

random_index = np.random.permutation(np.arange(n_samples))
X = X[random_index]
y = y[random_index]

#テストデータの分割
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

＜ロジスティック回帰＞

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

#パイプラインの作成
pipe_logistic = Pipeline([('scl',StandardScaler()),('est',LogisticRegression(random_state=1))])
pipe_logistic.fit(X_train,y_train)

#予測する
y_pred = pipe_logistic.predict(X_test)

#スコアを算出
print("accuracy_score:",accuracy_score(y_test, y_pred))
print("precision_score:",precision_score(y_test, y_pred))
print("recall_score",recall_score(y_test, y_pred))
print("f1_score",f1_score(y_test, y_pred))
```

```python
#可視化
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X_test, np.array(y_test), clf=pipe_logistic,res=0.02, legend=2)
```

＜ＳＶＭ＞

```python
from sklearn.svm import SVC

#パイプラインの作成
pipe_svc = Pipeline([('scl',StandardScaler()),('est',SVC(random_state=1))])
pipe_svc.fit(X_train,y_train)


#予測する
y_pred = pipe_svc.predict(X_test)

#スコアを算出
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
print("accuracy_score:",accuracy_score(y_test, y_pred))
print("precision_score:",precision_score(y_test, y_pred))
print("recall_score",recall_score(y_test, y_pred))
print("f1_score",f1_score(y_test, y_pred))
```

```python
#可視化
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X_test, np.array(y_test), clf=pipe_svc,res=0.02, legend=2)
```

＜決定木＞

```python
from sklearn.tree import DecisionTreeClassifier

#パイプライン作成
pipe_tree = Pipeline([('scl',StandardScaler()),('est',DecisionTreeClassifier(random_state=1))])
pipe_tree.fit(X_train,y_train)


#予測する
y_pred = pipe_tree.predict(X_test)

#スコアを算出
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
print("accuracy_score:",accuracy_score(y_test, y_pred))
print("precision_score:",precision_score(y_test, y_pred))
print("recall_score",recall_score(y_test, y_pred))
print("f1_score",f1_score(y_test, y_pred))
```
