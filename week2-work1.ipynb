# 【課題】統計基礎用語の説明

標準偏差とはデータの散らばり具合を示す指標である。
データセットのある指標がどの程度散らばっているかを示す。
分散は標準偏差の2乗であるため似た指標であるが、標準偏差は単位の次元が現実のデータと同じである。


共分散行列とは多次元データの共分散を表した行列である。
n個のm次元データがあった時、m個の変数それぞれのペアについて共分散m×m行列となる。

1次元の正規分布とは平均値と中央値が一致するデータの分布を表しており、平均値を中心にして左右対称に釣鐘型でデータが散らばっている。




 （練習）平均10、標準偏差5、生成乱数の数10000で1次元正規分布を作成。

```python
from numpy.random import *
from matplotlib import pyplot as plt
 
# 平均10, 標準偏差5の正規乱数を10万個生成
values = normal(10, 5, 1000000)


#values = normal(loc, scale, size)
#loc: 生成したい正規分布の平均を指定(default=0)
#scale: 生成したい正規分布の標準偏差を指定(default=1)
#size: 生成する乱数の数を指定(default=None)

# ヒストグラム
plt.hist(values, bins=50)
plt.show()
```

# 【課題】2次元正規分布データ点の作成

```python
import numpy as np
from numpy.random import *
from matplotlib import pyplot as plt
import seaborn as sns
 
mu_data1 = [-3, 0]
mu_data2 = [0, -3]
sigma = [[1.0, 0.8], [0.8, 1.0]]
 
# 2次元正規乱数を1万個生成
data1 = multivariate_normal(mu_data1, sigma, 500)
data2 = multivariate_normal(mu_data2, sigma, 500)


# 散布図
sns.jointplot(data1[:,0], data1[:,1])
plt.show()

sns.jointplot(data2[:,0], data2[:,1])
plt.show()
```

# 【課題】データ点へのラベル付け

```python
data1_ravel = np.full(500,0)
data2_ravel = np.full(500,1)

data1_raveled_0 = np.append(data1,data1_ravel.reshape(len(data1_ravel),1), axis=1)
data1_raveled_1 = np.append(data2,data2_ravel.reshape(len(data2_ravel),1), axis=1)


concat_data1_data2 = np.concatenate([data1_raveled_0,data1_raveled_1])
```

```python
concat_data1_data2
```

# 【課題】正しくラベル付けされているか見てみる

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(concat_data1_data2[:500,0],concat_data1_data2[:500,1], c='yellow', s=50, marker='o')
plt.scatter(concat_data1_data2[500:,0],concat_data1_data2[500:,1], c='purple', s=50, marker='o')
plt.show()
```
