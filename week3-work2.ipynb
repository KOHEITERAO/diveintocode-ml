## 【課題5】用語の調査

尖度とは正規分布と比べてどれだけとがっているかを表す統計量である。つまり山の尖り度と裾の拡がり度を示している。
正規分布の場合0となり、それ以上に尖りが大きいと正の値、逆の場合負の値をとる。

歪度とは正規分布に対しどれだけゆう画んでいるかを表す統計量である。つまり、左右対称性を示す指標。
正規分布の場合0となり、右裾が長いと正の値、左裾が長いと負の値をとる。
